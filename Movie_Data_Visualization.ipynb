# Top 5






```python
# Defining the color palette

BLUE, BLUE2, BLUE3, BLUE4, BLUE5 = '#03045e', '#0077b6', "#00b4d8", '#90e0ef', '#CDDBF3'
GRAY1, GRAY2, GRAY3, GRAY4, GRAY5 = '#212529', '#495057', '#adb5bd', '#dee2e6', '#f8f9fa'
RED1, ORANGE1, YELLOW1, GREEN1, GREEN2 = '#e76f51', '#f4a261',	'#e9c46a', '#4c956c', '#2a9d8f'
```

```python
top = df.head().copy()
pd.options.display.float_format = '${:,.0f}'.format # Formatting the Display of Numbers
top
```

```python
# Creating Bar Graph

fig, ax = plt.subplots(figsize=(10,4))
sns.set_theme(style='whitegrid')

ax = sns.barplot(data=top, x=top.FILM,
                 y=top.PROFIT, palette=[BLUE2])


ax.set_title('Top 5 highest-grossing movies', loc='left', fontsize=19,
             color = GRAY1)
ax.set_xlabel('Movies', fontsize=13, color= GRAY3)
ax.set_ylabel('PROFIT\n (in Billion)', fontsize=13,color= GRAY3)

ax.xaxis.set_tick_params(labelsize=14, labelcolor=GRAY2)
sns.despine(left=True, bottom=True)

formatted_labels = [film.replace(":", ":\n").replace("-", "-\n") for film in df.head(5)['FILM']]
ax.set_xticklabels(formatted_labels, rotation=0, ha='center')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: '{:,.0f}'.format(x/1000000)))


plt.show()
```

# EXPLORING BOX OFFICE PERFORMANCE OVER TIME FOR TOP FILMS

```python
# Creating the data frame

df_line= df.groupby(by='YEAR')['PROFIT'].sum()
df_line = pd.DataFrame(df_line)
df_line = df_line.reset_index()
idx = df.groupby('YEAR')['BOX_OFFICE'].idxmax()
df_line = df_line.merge(df.loc[idx, ['YEAR', 'FILM', 'BOX_OFFICE']],
                        on='YEAR', how='left')

df_line.rename(columns={'PROFIT':'PROFIT_YEAR','FILM':'TOP_FILM_YEAR'}, inplace = True)
pd.options.display.float_format = '${:,.0f}'.format # Formatting the Display of Numbers
```

```python
df_line
```

```python
df_line_film = df_line.copy()
```

```python
df_line_film.sort_values(by='PROFIT_YEAR', ascending = False, inplace = True)
```

```python
# Formatting numeric columns

def FORMATTING_PROFIT(valor):
    return f'{valor / 1e9:.2f}B'

df_line_film['PROFIT_YEAR_FORMATTED'] = df_line_film['PROFIT_YEAR'].apply(FORMATTING_PROFIT)
df_line_film['BOX_OFFICE_FOMATTED'] = df_line_film['BOX_OFFICE'].apply(FORMATTING_PROFIT)
```

```python
df_line_film.drop(columns={'PROFIT_YEAR','BOX_OFFICE'},inplace=True)
```

```python
df_line_film
```

```python
marker_colors = ['red', 'blue', 'green', 'yellow', 'black', 'orange', 'brown', 'gray']


sns.set_theme(style='whitegrid')

plt.figure(figsize=(12, 6))

sns.lineplot(data=df_line_film.head(8), x='YEAR', y='BOX_OFFICE_FOMATTED',
             hue='TOP_FILM_YEAR', marker='o',
             markersize=12,palette=marker_colors)

plt.title('Total Profit per Year for Top Grossing Films')
plt.xlabel('Year')
plt.ylabel('Profit (in billions)')

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

ax.yaxis.set_tick_params(labelsize=10,labelcolor=GRAY2)
sns.despine(left=True,bottom=True)


plt.show()
```

# Creating pie chart



```python
df_line_film= df_line_film.reset_index()
```

```python
df_line_film
```

```python
df_pie= df_line_film[['YEAR','PROFIT_YEAR_FORMATTED','TOP_FILM_YEAR']].copy()

df_pie['PROFIT_YEAR_FORMATTED'] = df_pie['PROFIT_YEAR_FORMATTED'].str.replace('B', '').astype(float)

total_profit = df_pie['PROFIT_YEAR_FORMATTED'].sum()

top_years = df_pie.nlargest(5,'PROFIT_YEAR_FORMATTED')

top_years['PERCENTAGE'] = (top_years['PROFIT_YEAR_FORMATTED']/total_profit)*100

plt.figure(figsize=(8,8))
plt.pie(top_years['PERCENTAGE'],
        labels=top_years['YEAR'],
        autopct='%1.1f%%',
        startangle=90,
        colors=['Salmon', 'SkyBlue', 'LightGreen', 'PeachPuff', 'Lavender'])
plt.title('Top 5 Profitable Years Contribution to Total Profit')

plt.show()
```
